```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def sir_rhs(t, y, params):
    """
    t    - timpul (nu îl folosim explicit, dar îl păstrăm în semnătură)
    y    - vectorul [S, I, R]
    params - dicționar cu parametrii modelului: {'beta': ..., 'gamma': ..., 'N': ...}
    """
    S, I, R = y
    beta = params['beta']
    gamma = params['gamma']
    N = params['N']

    dSdt = -beta * S * I / N
    dIdt =  beta * S * I / N - gamma * I
    dRdt =  gamma * I

    return np.array([dSdt, dIdt, dRdt], dtype=float)
```

```python
def euler_step(f, t_n, y_n, h, params):
    """
    Un singur pas de Euler explicit:
    y_{n+1} = y_n + h * f(t_n, y_n)
    """
    return y_n + h * f(t_n, y_n, params)


def rk4_step(f, t_n, y_n, h, params):
    """
    Un singur pas de metoda Runge-Kutta de ordin 4 (clasică).
    """
    k1 = f(t_n,         y_n,                 params)
    k2 = f(t_n + h/2.0, y_n + h * k1 / 2.0, params)
    k3 = f(t_n + h/2.0, y_n + h * k2 / 2.0, params)
    k4 = f(t_n + h,     y_n + h * k3,       params)

    return y_n + (h/6.0) * (k1 + 2*k2 + 2*k3 + k4)


def solve_ode(f, t0, t_end, y0, h, params, method='euler'):
    """
    Solver generic cu pas constant pentru un sistem de EDO:
    f  - funcția f(t, y, params)
    [t0, t_end] - intervalul de timp
    y0 - condiție inițială (vector numpy)
    h  - pasul de timp
    method - 'euler' sau 'rk4'
    """
    if method not in ['euler', 'rk4']:
        raise ValueError("method trebuie sa fie 'euler' sau 'rk4'")

    n_steps = int(np.ceil((t_end - t0) / h))
    t_values = np.zeros(n_steps + 1)
    y_values = np.zeros((n_steps + 1, len(y0)))

    t_values[0] = t0
    y_values[0] = y0

    t = t0
    y = y0.copy()

    for i in range(1, n_steps + 1):
        if method == 'euler':
            y = euler_step(f, t, y, h, params)
        else:
            y = rk4_step(f, t, y, h, params)

        t = t0 + i * h
        t_values[i] = t
        y_values[i] = y

    return t_values, y_values
```

```python
# Parametrii de bază
params = {
    'beta': 0.3,
    'gamma': 0.1,
    'N': 1000
}

# Condiții inițiale
S0 = 999
I0 = 1
R0 = 0
y0 = np.array([S0, I0, R0], dtype=float)

t0 = 0.0
t_end = 160.0
h = 0.1  # pasul de timp (poți schimba)

# Euler
t_euler, y_euler = solve_ode(sir_rhs, t0, t_end, y0, h, params, method='euler')

# RK4
t_rk4, y_rk4 = solve_ode(sir_rhs, t0, t_end, y0, h, params, method='rk4')

# Plot comparație
plt.figure(figsize=(12, 6))

# Doar populatia infectată, ca să vezi clar diferența
plt.plot(t_euler, y_euler[:, 1], label='I(t) - Euler', linestyle='--')
plt.plot(t_rk4, y_rk4[:, 1], label='I(t) - RK4')

plt.xlabel('Timp')
plt.ylabel('Numar persoane')
plt.title('Model SIR - comparatie Euler vs RK4 (h = {:.2f})'.format(h))
plt.legend()
plt.grid(True)
plt.show()

# Plot toate componentele pentru RK4 (soluția "mai bună")
plt.figure(figsize=(12, 6))
plt.plot(t_rk4, y_rk4[:, 0], label='S(t)')
plt.plot(t_rk4, y_rk4[:, 1], label='I(t)')
plt.plot(t_rk4, y_rk4[:, 2], label='R(t)')
plt.xlabel('Timp')
plt.ylabel('Numar persoane')
plt.title('Model SIR - solutie cu RK4')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

initial_infected = [1, 10, 50]

for I0_val in initial_infected:
    y0_pert = np.array([params['N'] - I0_val, I0_val, 0], dtype=float)
    t_rk4, y_rk4 = solve_ode(sir_rhs, t0, t_end, y0_pert, h, params, method='rk4')
    plt.plot(t_rk4, y_rk4[:, 1], label=f'I(0) = {I0_val}')

plt.xlabel('Timp')
plt.ylabel('I(t)')
plt.title('Perturbarea conditiilor initiale (I(0))')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

beta_values = [0.1, 0.3, 0.5]

for beta_val in beta_values:
    params_beta = params.copy()
    params_beta['beta'] = beta_val
    t_rk4, y_rk4 = solve_ode(sir_rhs, t0, t_end, y0, h, params_beta, method='rk4')
    plt.plot(t_rk4, y_rk4[:, 1], label=f'beta = {beta_val}')

plt.xlabel('Timp')
plt.ylabel('I(t)')
plt.title('Influenta parametrului beta (rata de transmitere)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

gamma_values = [0.05, 0.1, 0.2]

for gamma_val in gamma_values:
    params_gamma = params.copy()
    params_gamma['gamma'] = gamma_val
    t_rk4, y_rk4 = solve_ode(sir_rhs, t0, t_end, y0, h, params_gamma, method='rk4')
    plt.plot(t_rk4, y_rk4[:, 1], label=f'gamma = {gamma_val}')

plt.xlabel('Timp')
plt.ylabel('I(t)')
plt.title('Influenta parametrului gamma (rata de recuperare)')
plt.legend()
plt.grid(True)
plt.show()
```
